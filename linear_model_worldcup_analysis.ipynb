```python
# Is obiusly that with more match played (MP) more minutes per match (Mins_Per_90_Playing_Time). Let's plot it
import matplotlib.pyplot as plt
plt.scatter(data_cleaned['MP'], data_cleaned['Mins_Per_90_Playing_Time'], alpha = 0.2)
plt.xlabel('Number of matches')
plt.ylabel('Minutes played per 90 playing time')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x = data_cleaned['MP'])
plt.show()
```

```python
sns.boxplot(x = data_cleaned['Mins_Per_90_Playing_Time'])
plt.show()
```

```python
# There is a linear behavior, but we can see a constant behavior in 0 number of matches and this can change our linear regression model
X = data_cleaned[data_cleaned['MP'] > data_cleaned['MP'].quantile(0.1)]['MP']
y = data_cleaned[data_cleaned['MP'] > data_cleaned['MP'].quantile(0.1)]['Mins_Per_90_Playing_Time']
```

```python
plt.scatter(X, y, alpha = 0.2)
plt.xlabel('Number of matches [MP]')
plt.ylabel('Minutes played per 90 playing time')
plt.grid('On')
plt.show()
```

```python
# Now, we have a scatter plot with an apparently linear behavior
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
plt.scatter(X_train, y_train, alpha = 0.2, label = 'TRAIN')
plt.scatter(X_test, y_test, alpha = 0.2, label = 'TEST', c = 'r', s = 8)
plt.xlabel('Number of matches [MP]')
plt.ylabel('Minutes played per 90 playing time')
plt.grid('On')
plt.legend()
plt.show()
```

```python
# Let's model a linear regression
from sklearn.linear_model import LinearRegression
lnr_rgr = LinearRegression()
minutes_played_per_game = lnr_rgr.fit(X_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1))
```

```python
# Check the coeficient and the intercep
minutes_played_per_game.coef_, minutes_played_per_game.intercept_
```

```python
X_train = X_train.values.reshape(-1, 1)
y_train = y_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
```

```python
# Let's predict all test data
y_predict = minutes_played_per_game.predict(X_test)
```

```python
# Let's see the differences between the target data set and target predict
plt.scatter(X_test, y_test, alpha = 0.2, label = 'TEST')
plt.scatter(X_test, y_predict, alpha = 0.2, s = 10, label = ' PREDICT')
plt.legend()
plt.grid('On')
plt.xlabel('Number of matches [MP]')
plt.ylabel('Minutes played per 90 playing time')
plt.show()
```

```python
# Create a linear function
import numpy as np

x = np.array([[0],
             [20]])
y = np.dot(x, minutes_played_per_game.coef_) + minutes_played_per_game.intercept_
```

```python
# Let's see the linear model
plt.scatter(X_test, y_test, alpha = 0.2, label = 'TEST DATA')
plt.plot(x, y, c = 'r', label = 'Linear model')
plt.legend()
plt.grid('On')
plt.xlabel('Number of matches [MP]')
plt.ylabel('Minutes played per 90 playing time')
plt.title('Linear model of Number of matches [MP] VS. Minutes played per 90 playing time')
plt.show()
```
