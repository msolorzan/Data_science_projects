```python
# import libraries
import pandas as pd
import numpy as np

dataset = pd.read_csv('/content/1-World_cup2022.csv')
```

```python
# See the first 10 elements
dataset.head(10)
```

```python
# See types of data on each column
dataset.info()
```

```python
# Check the uniques values
dataset['Pos'].unique()
```

```python
position = dataset['Pos'].str.split(pat = ',', expand = True)
position
```

```python
# Create a copy to modify and not affect the original
dataset_copy = dataset.copy()
```

```python
dataset_copy.drop(['Pos'], axis = 1, inplace = True)
```

```python
dataset_copy.head()
```

```python
# rename columns header
position.rename(columns = {0 : 'pos_1', 1: 'pos_2'}, inplace = True)
```

```python
# Concatenate positions and dataset_copy
data = pd.concat([dataset_copy, position], axis = 1)
data
```

```python
data.info()
```

```python
data['MP_Playing_Time'].unique()
```

```python
data_not_nan = data.loc[:, 'Min_Playing_Time':'MP_Playing_Time'].fillna(value = 0.)
```

```python
data.drop(list(data.loc[:, 'Min_Playing_Time':'MP_Playing_Time'].columns), axis = 1,inplace = True)
```

```python
data
```

```python
data = pd.concat([data, data_not_nan], axis = 1)
```

```python
data.info()
```
